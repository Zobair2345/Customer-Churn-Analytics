```python
plt.figure(figsize=(8, 5))

df.boxplot(
    column="MonthlyCharges",
    by="Churn"
)

plt.title("Monthly Charges by Churn Status")
plt.suptitle("")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")

plt.show()
```

## Business Interpretation

Customers who churn generally have higher monthly charges. The median monthly charge for churned customers is approximately $80, compared to approximately $64 for retained customers.

However, the distributions overlap considerably, indicating that monthly charges alone are not sufficient to explain customer churn.

## Business Recommendation

TeleConnect should investigate whether customers on higher-priced plans perceive sufficient value for their monthly costs. Retention strategies could include plan reviews, personalised offers, and value-focused service bundles for high-charge customers.

```python
plt.figure(figsize=(8, 5))

df.boxplot(
    column="tenure",
    by="Churn"
)

plt.title("Customer Tenure by Churn Status")
plt.suptitle("")
plt.xlabel("Churn")
plt.ylabel("Tenure (Months)")

plt.show()
```

## Business Interpretation

Customers who churn generally have significantly shorter tenure. The median tenure of churned customers is approximately 10 months, compared to approximately 38 months for retained customers.

This indicates that newer customers are at a substantially higher risk of churn.

## Business Recommendation

TeleConnect should prioritise customer retention during the early stages of the customer lifecycle. Improved onboarding, proactive support, and targeted loyalty incentives during the first year may help reduce churn.

1. **Contract Type**

3. **Dependents**
   - Customers without dependents have a higher churn rate.

   - Gender does not appear to be a strong churn indicator.

   - Fiber Optic customers have the highest churn rate at approximately 42%.

6. **Tech Support**
   - Customers without Tech Support have a substantially higher churn rate.

7. **Online Security**
   - Customers without Online Security have a churn rate of approximately 42%.

8. **Online Backup and Device Protection**
   - Customers without these services show higher churn rates.

9. **Streaming Services**

   - These do not appear to be strong standalone churn indicators.

10. **Payment Method**
    - Electronic Check customers have the highest churn rate at approximately 45%.

11. **Monthly Charges**
    - Churned customers generally have higher monthly charges.

12. **Tenure**
