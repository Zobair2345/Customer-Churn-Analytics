# Sprint 3 - Data Cleaning

## Objective

The objective of this notebook is to clean the raw customer churn dataset by identifying and correcting data quality issues, preparing the data for exploratory data analysis (EDA), machine learning, and business reporting.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df_clean = df.copy()
```

```python
df_clean.info()
```

```python
(df_clean["TotalCharges"] == " ").sum()
```

```python
#replace the blank with nan
df_clean["TotalCharges"] = df_clean["TotalCharges"].replace(" ", np.nan)
```

```python
df_clean.isnull().sum()
```

Because I  already investigated the data and found those customers all have:

tenure = 0
They are brand new customers
They haven't accumulated any charges yet

This is a business-driven decision, not just a coding decision.

```python
df_clean["TotalCharges"] = df_clean["TotalCharges"].fillna(0)
```

```python
#converting to numerical
df_clean["TotalCharges"] = pd.to_numeric(df_clean["TotalCharges"])
```

```python
df_clean.info()
```

investigated the affected records and found they all had a tenure of 0 months, indicating they were new customers who had not yet accumulated any charges. Replacing the missing values with 0 preserved valid customer records while accurately reflecting the business context.

```python
df_clean.duplicated().sum()
```
