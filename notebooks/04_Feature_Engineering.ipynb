## Objective

The objective of this notebook is to transform the cleaned customer data into a format suitable for machine learning.

This includes removing non-predictive identifiers, separating features and the target variable, encoding categorical variables, and preparing numerical features.

```python
import pandas as pd
import numpy as np

df = pd.read_csv(
    "../data/processed/telco_churn_clean.csv"
)

df.head()
```

```python
df = df.drop("customerID", axis=1)

df.head()
```

## Feature Selection – Customer ID

The `customerID` column was removed because it is a unique identifier and does not provide meaningful behavioural information for predicting customer churn.

```python
df.shape
```

```python
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
#find categorical column
categorical_columns = X.select_dtypes(
    include=["object", "str"]
).columns

categorical_columns
```

```python
X_encoded = pd.get_dummies(
    X,
    columns=categorical_columns,
    drop_first=True,
    dtype=int
)

X_encoded.head()
```

```python
X_encoded.shape
X_encoded.columns
```

```python
y_encoded = y.map({
    "No": 0,
    "Yes": 1
})

y_encoded.value_counts()
```

```python
## Target Encoding

The target variable `Churn` was converted from categorical labels to binary numerical values.

- `0` represents customers who did not churn.
- `1` represents customers who churned.

Binary encoding prepares the target variable for classification algorithms.
```

```python
y_encoded = y.map({
    "No": 0,
    "Yes": 1
})

y_encoded.value_counts()
```

## Target Encoding

The target variable `Churn` was converted from categorical labels to binary numerical values.

- `0` represents customers who did not churn.
- `1` represents customers who churned.

Binary encoding prepares the target variable for classification algorithms.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

## Train-Test Split

The dataset was divided into 80% training data and 20% testing data.

Stratified sampling was used to preserve the original churn class distribution in both datasets. A fixed random state was used to ensure reproducibility.

```python
from  sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

numerical_columns = [
    "tenure",
    "MonthlyCharges",
    "TotalCharges"

]
```

```python
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numerical_columns] = scaler.fit_transform(
    X_train[numerical_columns]
)

X_test_scaled[numerical_columns] = scaler.transform(
    X_test[numerical_columns]
)
```

```python
X_train_scaled[numerical_columns].describe().round(2)
```

# Feature Engineering Summary

The customer churn dataset was prepared for machine learning by:

- Removing the non-predictive `customerID` identifier.
- Separating input features (`X`) and the target variable (`y`).
- Applying one-hot encoding to categorical variables.
- Encoding churn as a binary target (0 = No Churn, 1 = Churn).
- Splitting the dataset into 80% training and 20% testing data using stratified sampling.
- Standardising numerical features using training data only to prevent data leakage.

The processed training and testing datasets are now ready for machine learning.
